**Term Frequency–Inverse Document Frequency (TF-IDF)** is a numerical statistic that reflects how important a word is to a document in a collection. It is often used as a weighting factor in information retrieval and text mining.

**Term Frequency (TF):** This measures how frequently a term occurs in a document. Since every document is different in length, it is possible that a term would appear much more times in long documents than shorter ones. Thus, the term frequency is often divided by the document length (the total number of terms in the document) as a way of normalization.

**Inverse Document Frequency (IDF):** This measures how important a term is. While computing TF, all terms are considered equally important. However, certain terms, such as "is", "of", and "that", may appear a lot of times but have little importance. Thus we need to weigh down the frequent terms while scaling up the rare ones, by computing the following: IDF(t) = log_e(Total number of documents / Number of documents with term t in it).

The min_df and max_df parameters are used to filter terms based on their document frequency (DF).

**min_df** specifies the minimum number of documents that a term must appear in to be included in the vocabulary, and can be either an integer (absolute count) or a float (percentage of the total number of documents).

**max_df** specifies the maximum number of documents a term can appear in. If it appears in more than this threshold, it is considered too common and is not included in the vocabulary.

Adjusting these parameters can help remove terms that are too rare (possibly typos or very uncommon terms) or too common (stop words that are not useful for analysis)

By applying the TfidfVectorizer with min_df=3 and max_df=0.95, it reduced the feature space from 35,199 unique terms to 15967 unique terms. This is a significant reduction and it shows that:

There were many terms that appeared in 3 or fewer documents, which now been filtered out because they were too rare (likely not informative enough for most analysis tasks).
There may have been a few terms that appeared in more than 95% of the documents, which were also removed because they were too common (likely stop words or very common words that don't contribute much to distinguishing between documents).

````python
# Initialize TfidfVectorizer with unigrams and bigrams
tfidf_vectorizer = TfidfVectorizer(ngram_range=(1, 2), min_df=3, max_df=0.95)

# Fit and transform the training data
tfidf_train = tfidf_vectorizer.fit_transform(docs_train)

# Check the shape of the resulting TF-IDF matrix
print("Shape of TF-IDF matrix with unigrams and bigrams:", tfidf_train.shape)
````

The **ngram_range** parameter of TfidfVectorizer determines the different n-gram sizes that will be included when converting text data into a feature matrix.

For example:

ngram_range=(1, 1) means only unigrams (single words) are considered. This is the default setting.

ngram_range=(1, 2) means both unigrams and bigrams (pairs of consecutive words) are considered.

ngram_range=(2, 2) means only bigrams are considered.

ngram_range=(1, 3) means unigrams, bigrams, and trigrams (sequences of three consecutive words) are all considered.

However, increasing the ngram_range can significantly increase the number of features, as each combination of words within the specified n-gram size becomes a separate feature. This can lead to higher computational costs and more memory usage.

the feature space has expanded to 64,084 unique terms. This expansion is expected because bigrams take into account pairs of consecutive words, which increases the vocabulary size dramatically.

````python
from sklearn.metrics import accuracy_score, f1_score

 # Define a custom Dataset class that inherits from PyTorch's Dataset class
class MovieReviewsDataset(torch.utils.data.Dataset):
    # Constructor for the dataset class.
    # encodings: Tokenized and encoded text data.
    # labels: Corresponding labels for the text data.
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    # Method to retrieve an individual item from the dataset.
    # idx: Index of the item to retrieve.
    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        item['labels'] = torch.tensor(self.labels[idx]) # Convert each encoding in the item to a PyTorch tensor.
        return item # Add the corresponding label as a tensor to the item.

    # Method to get the size of the dataset.
    def __len__(self):
        return len(self.labels)

train_dataset = MovieReviewsDataset(train_encodings, y_train)
test_dataset = MovieReviewsDataset(test_encodings, y_test)

# Function to compute the evaluation metrics.
def compute_metrics(pred):
    labels = pred.label_ids   # Extract the actual labels from the predictions object.
    preds = pred.predictions.argmax(-1)   # Get the index of the highest logit to get the most likely label.
    f1 = f1_score(labels, preds) # Calculate the F1 score.
    acc = accuracy_score(labels, preds)  # Calculate the accuracy.
    # Return the computed metrics.
    return {
        'accuracy': acc,
        'f1': f1,
    }
# The pre-trained model to be evaluated.
# Directory where any outputs can be saved.
# Batch size for evaluation per device.
# The function used to compute evaluation metrics.
# The test dataset for evaluation.
trainer = Trainer(
    model=model,
    args=TrainingArguments(
        output_dir="/tmp/results",
        per_device_eval_batch_size=64,
    ),
    compute_metrics=compute_metrics,
    eval_dataset=test_dataset
)

# Evaluate the model
eval_results = trainer.evaluate()
print(eval_results)
````

````python
import torch
from torch.utils.data import DataLoader, Dataset
from transformers import AutoModelForSequenceClassification, AutoTokenizer
from sklearn.datasets import load_files
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score

# Define the model name from Hugging Face and load the pre-trained model and tokenizer
model_name = "JamesH/autotrain-third-project-1883864250"
model = AutoModelForSequenceClassification.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Load the movie reviews dataset from a specified folder
movie_reviews_data_folder = "txt_sentoken"
dataset = load_files(movie_reviews_data_folder, shuffle=False, encoding='utf-8')

# Split the dataset into training and testing sets
docs_train, docs_test, y_train, y_test = train_test_split(dataset.data, dataset.target, test_size=0.25, random_state=None)

# Define a custom dataset class for handling the tokenized text data
class MovieReviewsDataset(Dataset):
    def __init__(self, texts, labels, tokenizer):
        self.encodings = tokenizer(texts, truncation=True, padding=True, max_length=512) # Tokenize the text data with truncation and padding
        self.labels = labels   # Store the labels

    # Method to retrieve an individual item from the dataset
    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}  # Create a dictionary for the encodings at the specified index
        item['labels'] = torch.tensor(self.labels[idx])   # Add the corresponding label to the dictionary
        return item

    # Method to get the size of the dataset
    def __len__(self):
        return len(self.labels)

# Create instances of the custom dataset for training and testing data
train_dataset = MovieReviewsDataset(docs_train, y_train, tokenizer)
test_dataset = MovieReviewsDataset(docs_test, y_test, tokenizer)

````

````python
# Importing the functional API from PyTorch's neural network module.
# This module contains many useful functions for building and training neural networks.
import torch.nn.functional as F

# Defining a function to compute the loss.
def compute_loss(predictions, labels):
    # Compute and return the cross-entropy loss.
    # Cross-entropy loss is commonly used in classification tasks.
    # 'predictions' are the logits (raw, unnormalized scores) outputted by the model.
    # 'labels' are the true labels of the data.
    # F.cross_entropy combines LogSoftmax and NLLLoss (Negative Log Likelihood Loss) in one single class.
    # It's useful when training a classification problem with C classes.
    # The function applies a softmax to the model's output and then computes the cross-entropy between the result and the true labels.
    return F.cross_entropy(predictions, labels)
````

````python
# Check if a GPU is available and set PyTorch to use the GPU
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Move the model to the appropriate device
model.to(device)
````

````python
# Define a function for evaluating the model
def evaluate_model(model, dataset):
    # DataLoader loads the dataset and provides batches of data for evaluation.
    # batch_size=16 defines the number of samples per batch.
    # shuffle=False ensures that the evaluation data is not shuffled, which is standard for evaluation.
    loader = DataLoader(dataset, batch_size=16, shuffle=False)

    # Set the model to evaluation mode. This turns off specific layers/processes like dropout.
    model.eval()

    # Initialize a variable to accumulate the total loss.
    total_loss = 0

    # Lists to store predictions and true labels for later calculation of metrics.
    predictions, true_labels = [], []

    # Context manager that disables gradient calculation.
    # Gradients are not needed for evaluation and it saves memory and computation.
    with torch.no_grad():
        for batch in loader:
            # Prepare input data and labels by moving them to the same device (GPU/CPU) as the model.
            inputs = {key: val.to(device) for key, val in batch.items() if key != 'labels'}
            labels = batch['labels'].to(device)

            # Forward pass: compute predicted outputs by passing inputs to the model.
            outputs = model(**inputs)

            # Extract logits (raw scores) from the model's output.
            logits = outputs.logits

            # Compute the loss for the current batch and accumulate it.
            loss = F.cross_entropy(logits, labels)
            total_loss += loss.item()

            # Convert logits to actual predictions (class with highest score).
            preds = torch.argmax(logits, dim=1)

            # Store predictions and true labels for calculating accuracy and F1 score.
            predictions.extend(preds.cpu().numpy())
            true_labels.extend(labels.cpu().numpy())

    # Calculate the average loss over all batches.
    average_loss = total_loss / len(loader)

    # Calculate accuracy and F1 score using the stored predictions and true labels.
    accuracy = accuracy_score(true_labels, predictions)
    f1 = f1_score(true_labels, predictions)

    # Return the calculated metrics.
    return average_loss, accuracy, f1

# Evaluate the model on the test dataset and print the results.
loss, accuracy, f1 = evaluate_model(model, test_dataset)
print(f"Loss: {loss:.4f}")
print(f"Accuracy: {accuracy:.4f}")
print(f"F1 Score: {f1:.4f}")

````

````python
def predict_with_huggingface_model(model, dataset):
    # DataLoader batches the dataset and provides an iterator over it
    loader = DataLoader(dataset, batch_size=16, shuffle=False)

    # Set the model to evaluation mode. This turns off dropout layers, etc.
    model.eval()

    # Container for storing all predictions
    predictions = []

    # Disables gradient calculations, saving memory and computations
    with torch.no_grad():
        for batch in loader:
            # Prepare batch data (forward pass inputs) and send them to the same device as the model
            inputs = {key: val.to(device) for key, val in batch.items() if key != 'labels'}

            # Forward pass: compute the output of the model on the batch data
            outputs = model(**inputs)

            # Convert logits to actual predictions (index of highest score)
            preds = torch.argmax(outputs.logits, dim=1)

            # Extend the list of predictions with the current batch's predictions
            predictions.extend(preds.cpu().numpy())

    # Return the compiled list of predictions for the entire dataset
    return predictions

%timeit predict_with_huggingface_model(james_h_model, test_dataset)
````
